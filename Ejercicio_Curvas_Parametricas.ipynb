# Ejercicio sección 2.7: 
Demuestre que la curva de ecuación $r = \cos( k \theta)$ tiene $k$ pétalos si $k$ es impar y $2k$ pétalos si $k$ es par.

```
# k par

import matplotlib.pyplot as mplt
import numpy as np

ax = mplt.subplot(111, projection='polar')

k= 2

theta1 = np.linspace(0,2*np.pi,10000) # linspace divide el intervalo (0, 2pi) en 10000 partes iguales
  
r1 = np.cos(k*theta1)
  
mplt.polar(theta1, r1, label='$r=(\cos($' + str(k)  + '$\\theta))$')#, color= colores[k1]  )

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1)) # Con esta instrucción modifico el lugar a imprimir la leyenda
mplt.show()
```

```
import matplotlib.pyplot as mplt
import numpy as np

ax = mplt.subplot(111, projection='polar')

colores = ['b', 'g', 'r']#, 'c', 'm'] # Lista de colores disponibles: {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}

for k in range(2,8,2): # vamos a graficar para 5 valores diferentes de a
  k1=int((k-2)/2)
  theta1 = np.linspace(0,2*np.pi,10000) # linspace divide el intervalo (0, 2pi) en 10000 partes iguales
  
  r1 = np.cos(k*theta1)
  
  mplt.polar(theta1, r1, label='$r=(\cos($' + str(k)  + '$\\theta))$', color= colores[k1]  )
  
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1)) # Con esta instrucción modifico el lugar a imprimir la leyenda
mplt.show()
```

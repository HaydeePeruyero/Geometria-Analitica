# Series de Fourier

Las funciones periódicas en $[0, 2\pi ]$ se pueden aproximar tanto como queramos usando sumas de funciones trigonométricas.

Jean Baptiste Fourier (1768-1830) como parte de sus investigaciones sobre las propiedades del calor y su difusión usó *series* trigonométricas. A estas las  llamamos ahora *series de Fourier*:

$S(x) = (1/2) a_0 + \sum\limits_{n=1}^{\infty} (a_n \cos(n \, x) + b_n \sin(n \, x)) $

Veamos un ejemplo de una aproximación *parcial* a una función definida por pedazos ajenos y que no es continua.

Ahora veamos como se implementan estas funciones.

```
# Como siempre, empezamos por importar los *módulos* que usaremos
import sympy as sym
from sympy import symbols
from sympy.plotting import plot

# Declaramos la variable simbólica 'x', para que la use sympy
x = symbols('x')

# Inicializamos la variable 'y', empezando en cero
y = 0

# Recorremos un rango de valores para encontrar las series parciales
for i in range(1,10,2):
  y_x=sym.pi/4*(sym.sin(i*x)/i)
  y = y + y_x
  print(y)  
```

```
import sympy as sym
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')

# Fijamos una lista con los colores que usaremos para las gráficas que aproximan la función inicial (por sus siglas en inglés).

colores = ['b', 'g', 'r', 'c', 'm', 'y']

y = 0
k=-1
for i in range(1,12,2):
  k=k+1
  y_x=sym.pi/4*(sym.sin(i*x)/i)
  y = y + y_x
  # Ahora imprimimos las funciones usando 'plot(_)' en sympy
  plot(y, (x, -2*sym.pi, 2*sym.pi), line_color= colores[k]) 
```

Otro ejemplo, con las gráficas en el mismo eje, ahora usando matplotlib. 

```
import matplotlib.pyplot as mplt
import numpy as np

ax = mplt.subplot(111)

colores = ['b', 'g', 'r', 'c']#, 'm', 'y'] # Lista de colores disponibles: {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
y=1
i=-1
xx = np.arange(0,2*np.pi,0.01)
for k in range(1,8,2): # vamos a graficar para 5 valores diferentes de a
 
  xx = np.arange(0,2*np.pi,0.01) 
  i=i+1
  yy = np.pi/4*np.sin(k*xx)
  
  y = y + yy
  
  mplt.plot(xx, y, color= colores[i]  )
  
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1)) # Con esta instrucción modifico el lugar a imprimir la leyenda
mplt.show()
```
